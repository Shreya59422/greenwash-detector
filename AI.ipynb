```python
import gradio as gr

# Greenwashing keywords
buzzwords = [
    "eco-friendly", "green", "natural", "chemical free", "sustainable",
    "non-toxic", "environmentally safe", "100% natural", "planet friendly",
    "biodegradable", "eco safe"
]

# Trusted certifications
certifications = [
    "fsc", "ecocert", "iso", "energy star",
    "fairtrade", "cruelty free certified", "leaping bunny"
]

def analyze_text(text):
    text_lower = text.lower()

    found_buzz = [w for w in buzzwords if w in text_lower]
    found_certs = [c for c in certifications if c in text_lower]

    risk_score = min(len(found_buzz) * 12, 100)

    verdict = ""
    if risk_score > 60:
        verdict = "High risk of greenwashing"
    elif risk_score > 30:
        verdict = "Moderate risk – verify claims"
    else:
        verdict = "Low risk – seems more trustworthy"

    result = f"""
 Analysis Result

Suspicious terms found: {', '.join(found_buzz) if found_buzz else 'None'}
Valid certifications found: {', '.join(found_certs) if found_certs else 'None'}

 Greenwashing Risk Score: {risk_score}/100
Verdict: {verdict}

Tip: Look for real certifications like FSC, ISO 14001, ECOCERT, Fairtrade.
"""
    return result


demo = gr.Interface(
    fn=analyze_text,
    inputs=gr.Textbox(lines=6, placeholder="Paste product description here..."),
    outputs="text",
    title=" Greenwash Detector",
    description="AI-powered tool to detect misleading eco-friendly claims in product descriptions.",
)

demo.launch(share=True)
```
